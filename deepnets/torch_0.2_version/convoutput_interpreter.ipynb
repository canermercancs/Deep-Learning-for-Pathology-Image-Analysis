```python
import pickle
import os 
import matplotlib.pyplot as plt

page_size = 5
num_model = 6
mdl_name = f'/projects/Medical/DigiPath/caner/ConsROI_patches/page{page_size}/models/alexnet_model{num_model}'
```

```python
fname = os.path.join(mdl_name, 'summary')
with open(fname, 'rb') as f:
    params = pickle.load(f)
    
fname = os.path.join(mdl_name, 'epoch_stats')
with open(fname, 'rb') as f:
    stats = pickle.load(f)
```

```python
print('configuration:')
print(params)
phase = ['loss', 'accuracy']
for idx, p in enumerate(phase):
    plt.plot(stats[idx]['train'], label='train', c='orange')
    plt.plot(stats[idx]['val'],   label='val',   c='blue')
    plt.legend(['train', 'val'])
    plt.title(p)
    plt.show()    
```
