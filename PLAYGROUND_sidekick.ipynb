### Randomly making Hue changes to an ROI

```python
from deepnets.transform.transforms import RandomHueJitter
from deepnets.transform.transforms import RandomRGBJitter
huetf = RandomHueJitter(p=1, jitt=0.80)#(p=1, jitt=0.25)
rgbtf = RandomRGBJitter(p=1, jitt=10)#(p=1, jitt=20)
```

```python
roi = pimObj.ConsensusROIs[2]
roi_rgb = roi.getRGB()
roi_he  = roi.getHE()

# make the transformations...
roi_jitt_hue = huetf(roi_rgb)
roi_jitt_rgb = rgbtf(roi_jitt_hue)

# plt.subplot(121)
# plt.imshow(roi_rgb)
# plt.axis('off')
# plt.subplot(122)
# plt.imshow(roi_jitt_hue)
# plt.title('hue')
# plt.axis('off')
# plt.show()
plt.subplot(121)
plt.imshow(roi_rgb)
plt.axis('off')
plt.subplot(122)
plt.imshow(roi_jitt_rgb)
plt.title('rgb')
plt.axis('off')
plt.show()
```
